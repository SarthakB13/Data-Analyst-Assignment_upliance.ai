```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the data
Df = pd.read_excel("E:\Data Analysis Assignment\Data Analyst Intern Assignment - Excel.xlsx", sheet_name = None)
```

```python
# Reading different sheets
UserDetail = Df['UserDetails.csv']
CookingSessions = Df['CookingSessions.csv']
OrderDetails = Df['OrderDetails.csv']
```

```python
# Joining the different Tables
Df_test = pd.merge(OrderDetails, CookingSessions, on=['User ID', 'Session ID'])
```

```python
Df_test = Df_test.drop(['Dish Name_y', 'Meal Type_y'], axis=1)
```

```python
Df_test
```

```python
Merged_df = pd.merge(Df_test, UserDetail, on='User ID')
```

```python
Merged_df = Merged_df.drop('Registration Date', axis=1)
```

```python
Merged_df
```

```python
Merged_df.info()
```

```python
Merged_df.isnull().sum()
```

```python
Merged_df['Rating'].fillna(round(Merged_df['Rating'].mean(), 0), inplace = True)
```

```python
meal_counts = Merged_df['Meal Type_x'].value_counts()
meal_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Orders by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
revenue_by_time = Merged_df.groupby('Time of Day')['Amount (USD)'].sum()
revenue_by_time.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Revenue by Time of Day')
plt.ylabel('')
plt.show()
```

```python
location_counts = Merged_df['Location'].value_counts()
location_counts.plot(kind='bar', color='cyan', edgecolor='black')
plt.title('Users by Location')
plt.xlabel('Location')
plt.ylabel('Number of Times User ordered')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_duration = Merged_df.groupby('Meal Type_x')['Duration (mins)'].mean()
avg_duration.plot(kind='bar', color='purple', edgecolor='black')
plt.title('Average Duration by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Duration (mins)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(Merged_df['Duration (mins)'], Merged_df['Amount (USD)'], color='red', alpha=0.7)
plt.title('Duration vs. Amount')
plt.xlabel('Duration (mins)')
plt.ylabel('Amount (USD)')
plt.grid()
plt.show()
```

```python
favorite_meal_counts = Merged_df['Favorite Meal'].value_counts()
favorite_meal_counts.plot(kind='bar', color='pink', edgecolor='black')
plt.title('Favorite Meal Distribution')
plt.xlabel('Meal')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()
```

```python
dish_counts = Merged_df['Dish Name_x'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=dish_counts.index, y=dish_counts.values, palette='viridis')

plt.title('Most Popular Dishes', fontsize=16)
plt.xlabel('Dish Name', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
dish_count = Merged_df.groupby(['Meal Type_x', 'Dish Name_x']).size().reset_index(name='Count')

dish_count_sorted = dish_count.sort_values(by=['Meal Type_x', 'Count'], ascending=[True, False])

plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Dish Name_x', hue='Meal Type_x', data=dish_count_sorted, palette="Set2")

plt.title('Popular Dishes by Meal Type')
plt.xlabel('Number of Orders')
plt.ylabel('Dish Name')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.violinplot(data=Merged_df, x='Meal Type_x', y='Age', palette="Set2")

plt.title('Distribution of Age by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Age')

plt.tight_layout()
plt.show()
```
